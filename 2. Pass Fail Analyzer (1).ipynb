```python
def pass_fail_analyzer(marks):
    pass_count = 0
    fail_count = 0

    for mark in marks:
        if mark >= 50:
            pass_count += 1
        else:
            fail_count += 1

    print("Total Passed Students:", pass_count)
    print("Total Failed Students:", fail_count)


marks = [45, 78, 90, 33, 60]
pass_fail_analyzer(marks)
```
